#### In this first assignment I will create a basic network graph using the NetworkX pacakge for Python.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

#### Empty graph

```python
graph = nx.Graph()
```

#### add nodes to the graph

```python
nodes = ['Gabriel', 'Bill', 'Casey', 'Don', 
        'Ed', 'Francis', 'Holly', 'Ira', 'John']
graph.add_nodes_from(nodes)
```

#### add edges to the graph

```python
edges = [('Gabriel', 'Bill'),
         ('Gabriel', 'Casey'),
         ('Gabriel', 'Don'),
         ('Gabriel', 'Francis'),
         ('Bill', 'Don'),
         ('Bill', 'Kary'),
         ('Bill', 'Ed'),
         ('Casey', 'Francis'),
         ('Casey', 'Don'),
         ('Ed', 'Don'),
         ('Ed', 'Kary'),
         ('Don', 'Francis'),
         ('Don', 'Kary'),
         ('Francis','Kary'),
         ('Holly', 'Francis'),
         ('Holly', 'Kary'),
         ('Holly', 'Ira'),
         ('Ira','John' )
        ]
graph.add_edges_from(edges)
```

#### Degree of each of the vertices and connecting thegraph

```python
nx.degree(graph)
```

#### Plotting the graph 

```python
pos=nx.spring_layout(graph)
nx.draw(graph,pos)
labels=nx.draw_networkx_labels(graph,pos)
plt.show()
```
